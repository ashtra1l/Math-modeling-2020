```python
#Define parameters
Fo = 0.5
#Create grid
nt = 100
nx = 100
T = zeros([nt, nx])
#Set initial condition
T[0,:] = 1.
#Set boundary conditions
T[:,0] = 0. 
T[:,-1] = 1.
#Loop through solution
for n in arange(nt-1):
    T[n+1,1:-1] = Fo*T[n,2:] + (1 - 2*Fo)*T[n,1:-1] + Fo*T[n,0:-2]
```
